```python
#Values of heights
x = np.array([[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]])
#values of weight
y = np.array([63,81,56,91,47,57,76,72,62,48])
```

```python
plt.xlabel('Height')
plt.ylabel('weight')
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
# value of m
reg.coef_
```

```python
# value of c
reg.intercept_
```

```python
#step by step
```

```python
x
```

```python
y
```

```python
x.shape,y.shape
```

```python
#step 1 :Calculate the means of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
mean_x,mean_y
```

```python
x = x.flatten()
```

```python
#Step 2: Calculate the slope (m)
numerator =  np.sum((x-mean_x)*(y - mean_y))
denominator = np.sum((x-mean_x)**2)
slope = numerator / denominator
```

```python
slope
```

```python
#Step 3:Calculate the intercept(b)
intercept = mean_y - slope*mean_x
```

```python
intercept
```

```python
#print the slope and intercept
print("Slope(m):", slope)
print("Intercept(c):", intercept)
```

```python
# y = m * x + C
175*slope+intercept
```

```python
149*slope+intercept
```

```python
y_pred = x * slope + intercept
```

```python
y_pred
```

```python
plt.xlabel('Height')
plt.ylabel('weight')
plt.scatter(x,y, label = 'actual data',color = 'r')
plt.plot(x,y_pred, label = 'predicted data',color = 'g')
plt.legend()
```

```python
#mean absolute error
```

```python
y_pred = y_pred.flatten()
```

```python
y-y_pred
```

```python
np.abs(y-y_pred)
```

```python
np.mean(np.abs(y-y_pred))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y,y_pred)
```
